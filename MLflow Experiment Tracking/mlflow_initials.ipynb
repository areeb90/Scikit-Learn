```python
import mlflow

mlflow.set_tracking_uri(
    "sqlite:///C:/101 Mahcine Learning/Projects/Scikit-Learn/MLflow Experiment Tracking/mlflow.db"
)

experiments = mlflow.search_experiments()
for e in experiments:
    print(f"Experiment: {e.name} | ID: {e.experiment_id}")

runs = mlflow.search_runs(search_all_experiments=True)
print(f"\nTotal runs found: {len(runs)}")
print(runs[['run_id', 'tags.mlflow.runName', 'metrics.test_roc_auc']].to_string())
```

```python
import mlflow
import os

# Check where runs actually went — search all possible mlflow.db files
import glob

# Search entire project folder
search_root = r"C:\101 Mahcine Learning\Projects\Scikit-Learn"
db_files = glob.glob(f"{search_root}\\**\\mlflow.db", recursive=True)
mlruns_dirs = glob.glob(f"{search_root}\\**\\mlruns", recursive=True)

print("mlflow.db files found:")
for f in db_files:
    size = os.path.getsize(f)
    print(f"  {f}  ({size:,} bytes)")

print("\nmlruns/ folders found:")
for d in mlruns_dirs:
    print(f"  {d}")
```

```python
import os

mlruns_path = r"C:\101 Mahcine Learning\Projects\Scikit-Learn\mlruns"

for root, dirs, files in os.walk(mlruns_path):
    level = root.replace(mlruns_path, '').count(os.sep)
    indent = '  ' * level
    print(f"{indent}{os.path.basename(root)}/")
    if level < 3:  # only show files for first 3 levels
        for f in files:
            print(f"{indent}  {f}")
```
